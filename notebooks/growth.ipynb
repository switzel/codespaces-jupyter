```python
from matplotlib import pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider, RadioButtons,Layout
from matplotlib.patches import Circle
from scipy.special import gamma as scigamma
import numpy as np
%matplotlib notebook
```

```python
fig = plt.figure()
ax = fig.subplots()
@interact(R=FloatSlider(min=1,max=50,value=1,step=.01),
         scale=RadioButtons(options=('linear','log')))
def fcn_plot(R=10,lin=True,square=True,cube=True,exp=False,gamma=False,scale='linear'):
    ax.clear()
    x = np.arange(0,R,.01)
    ax.set_yscale(scale)
    if lin:
        ax.plot(x,x)
    if square:
        ax.plot(x,x*x)
    if cube:
        ax.plot(x,x*x*x)
    if exp:
        ax.plot(x,np.exp(x))
    if gamma:
        ax.plot(x,scigamma(x+1))
fcn_plot(10,exp=True,gamma=True)
```
