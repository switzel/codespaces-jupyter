```python
from matplotlib import pyplot as plt
from ipywidgets import interact, FloatSlider, IntSlider, Layout
from matplotlib.patches import Circle
import numpy as np
%matplotlib notebook
```

```python
fig = plt.figure()
ax = fig.subplots()
ax.set_xlim(-1.1,1.1)
ax.set_ylim(-1.1,1.1)
ax.set_aspect(1)
xs = [1,1,1,1,1,1,1]
ys = [0,0,0,0,0,0,0]
ax.add_patch(Circle((0,0),1,fill=False))
pts = ax.scatter(xs,ys,c=['k','b','g','y','r','c','m'])
@interact(x=FloatSlider(min=0,max=1,value=0,step=.01/15,layout=Layout(width='100%')),
          d=IntSlider(min=1,max=6,value=2))
def f(x,d):
    r = [d*x] + [x + i/d for i in range(d)]
    pts.set_offsets([[np.cos(x*2*np.pi),np.sin(x*2*np.pi)] for x in r])
    fig.canvas.draw_idle()
```
